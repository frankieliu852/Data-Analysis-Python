## Grouping with functions

```python
import pandas as pd
import numpy as np
```

```python
df3 = pd.DataFrame(np.random.randn(5,5), columns=['a','b','c','d','e'], index=['AA','BBB','CC','DDDDD','EEE'])
df3
```

```python
# Default is set to axis=0

df3.groupby(len).sum()
```

```python
df3.groupby(len, axis=1).sum()
```

```python
df3.groupby([len, df3.index]).sum()
```
